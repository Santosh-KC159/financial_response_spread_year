# Cross response functions and average log return by average trade sign plots

```python
import numpy as np
from matplotlib import pyplot as plt
import pickle

%matplotlib inline
```

```python
ticker_i = 'MSFT'
ticker_j = 'MSFT'
days = ['07','08','09','10','11']
tau_val = 1000
t_step = 100
```

```python
i = 0

plt.figure(figsize=(9,16))
plt.subplots_adjust(hspace = 0, wspace = 0)

for day in days:
    cross = pickle.load(open('../Data/cross_response_functions_data_%dms/cross_201603%s_%si_%sj.pickl' % (t_step, day, ticker_i, ticker_j), 'rb'))
    avg = pickle.load(open('../Data/avg_return_sign_data_%dms/avg_201603%s_%si_%sj.pickl' % (t_step, day, ticker_i, ticker_j), 'rb'))
    
    i += 1
    
    plt.subplot(5,2,i)
    plt.semilogx(cross, '-g', label='Stock i %s and stock j %s - %s' % (ticker_i, ticker_j, day))
    plt.xlabel(r'Time lag $[\tau]$')
    plt.ylabel(r'Cross response $ R_{ij} (\tau) $')
    plt.legend(loc='best')
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    plt.grid(True)
    plt.tight_layout()

    i += 1
    
    plt.subplot(5,2,i)
    plt.semilogx(avg, '-g', label='Stock i %s and stock j %s - %s' % (ticker_i, ticker_j, day))
    plt.xlabel(r'Time lag $[\tau]$')
    plt.ylabel(r'$ \left \langle r_{i}\left ( t, \tau \right ) \right \rangle \left \langle \epsilon_{j} \left ( t \right ) \right \rangle $')
    plt.legend(loc='best')
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    plt.grid(True)
    plt.tight_layout()
    
plt.savefig('cross_response_avg_comparison_%s_%s_%dms.png' % (ticker_i, ticker_j, t_step))
```
