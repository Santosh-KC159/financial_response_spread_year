# My results

```python
# Load data

ticker_i = 'AAPL'
ticker_j = 'MSFT'
__tau__ = 1000
months_list = []
days_list = []
year = 2008

for i in range(1,13):
    if (i < 10):
        months_list.append('0' + str(i))
    else:
        months_list.append(str(i))

for month in months_list:

    m = pickle.load(open('../../Basic/days_{}.pickle'.format(month), 'rb'))
    days_list += [m]
```

## Cross-response

```python
fig = plt.figure(figsize=(16, 9))

AAPL_MSFT_plot_data_year = np.zeros(__tau__)
MSFT_AAPL_plot_data_year = np.zeros(__tau__)

for month in months_list:

    AAPL_MSFT_plot_data_month = np.zeros(__tau__)
    MSFT_AAPL_plot_data_month = np.zeros(__tau__)

    for i, day in enumerate(days_list[int(month) - 1]):

        AAPL_MSFT_day = pickle.load(open(''.join((
            '../../Cross_response_individual_stock/taq_data_{2}/taq_cross_response_data/taq_cross'
            + '_response_data_{2}{3}{4}_{0}i_{1}j.pickle').split())
            .format(ticker_i, ticker_j, year, month, day), 'rb'))
        
        MSFT_AAPL_day = pickle.load(open(''.join((
            '../../Cross_response_individual_stock/taq_data_{2}/taq_cross_response_data/taq_cross'
            + '_response_data_{2}{3}{4}_{0}i_{1}j.pickle').split())
            .format(ticker_j, ticker_i, year, month, day), 'rb'))

        AAPL_MSFT_plot_data_month += AAPL_MSFT_day
        MSFT_AAPL_plot_data_month += MSFT_AAPL_day

    AAPL_MSFT_plot_data_month = AAPL_MSFT_plot_data_month / len(days_list[int(month) - 1])
    MSFT_AAPL_plot_data_month = MSFT_AAPL_plot_data_month / len(days_list[int(month) - 1])

    AAPL_MSFT_plot_data_year += AAPL_MSFT_plot_data_month
    MSFT_AAPL_plot_data_year += MSFT_AAPL_plot_data_month

AAPL_MSFT_plot_data_year = AAPL_MSFT_plot_data_year / len(months_list)
MSFT_AAPL_plot_data_year = MSFT_AAPL_plot_data_year / len(months_list)

plt.semilogx(AAPL_MSFT_plot_data_year, linewidth=5, label='AAPL - MSFT')
plt.semilogx(MSFT_AAPL_plot_data_year, linewidth=5, label='MSFT - AAPL')

plt.legend(loc='best', fontsize=25)
plt.title('Juan cross-response results', fontsize=32)
plt.xlabel(r'$\tau \, [s]$', fontsize=25)
plt.ylabel(r'$R_{ij}(\tau)$', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
#plt.ylim(2 * 10 ** -5, 12 * 10 ** -5)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
plt.grid(True)
plt.tight_layout()
```
