```python
# Modules

from matplotlib import pyplot as plt
import numpy as np

import pickle
```

## Load data

```python
R_AAP_MSFT = np.loadtxt('../Data/Detail/AAPL-2MSFT_2008_RDC_L=1000.txt',
                        usecols=1, unpack=True)
R_MSFT_AAPL = np.loadtxt('../Data/Detail/MSFT-2AAPL_2008_RDC_L=1000.txt',
                         usecols=1, unpack=True)
S_AAPL = np.loadtxt('../Data/Detail/AAPL-2AAPL_2008_RDC_L=1000.txt',
                    usecols=1, unpack=True)
S_MSFT = np.loadtxt('../Data/Detail/MSFT-2MSFT_2008_RDC_L=1000.txt',
                    usecols=1, unpack=True)
```

## Self-response

```python
fig = plt.figure(figsize=(16, 9))
plt.semilogx(S_AAPL, linewidth=5, label='AAPL')
plt.semilogx(S_MSFT, linewidth=5, label='MSFT')
plt.legend(loc='best', fontsize=25)
plt.title('S. Wang self-response results', fontsize=32)
plt.xlabel(r'$\tau \, [s]$', fontsize=25)
plt.ylabel(r'$R_{ii}(\tau)$', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.ylim(0.8 * 10 ** -4, 2.4 * 10 ** -4)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
plt.grid(True)
plt.tight_layout()
```

## Cross-response

```python
fig = plt.figure(figsize=(16, 9))
plt.semilogx(R_AAP_MSFT, linewidth=5, label='AAPL - MSFT')
plt.semilogx(R_MSFT_AAPL, linewidth=5, label='MSFT - AAPL')
plt.legend(loc='best', fontsize=25)
plt.title('S. Wang cross-response results', fontsize=32)
plt.xlabel(r'$\tau \, [s]$', fontsize=25)
plt.ylabel(r'$R_{ij}(\tau)$', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.ylim(2 * 10 ** -5, 9 * 10 ** -5)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
plt.grid(True)
plt.tight_layout()
```
